```python
# Importing the dependencies 

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Importing the dataset 

df = pd.read_csv('creditcard.csv')
df.head()
```

```python
# Number of fraud card :
#0 --> Normal Transaction
#1 --> Fraudulent transaction

df['Class'].value_counts()
```

```python
# Separating the data for analysis

legit = df[df.Class == 0]
fraudulent = df[df.Class == 1]
```

Under-Sampling

Build a sample dataset containing simillar distrubution of normal transaction and Fraudulan Transaction

Number of Fraudulent transaction --> 492

```python
# Take a random values on legit values
legit_sample = legit.sample(n=492)
```

```python
# Concatenating Two dataframe

new_df = pd.concat([legit_sample,fraudulent],axis=0)
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
# Splitting the data into features and targets 

X = new_df.drop('Class',axis=1)
y = new_df['Class']
```

```python
 X.head()
```

```python
y.head()
```

```python
# Split the data into Traning and Testing data

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2,test_size=0.2,stratify=y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Model training 

# Initialization
lg = LogisticRegression()

# Fit the data
lg.fit(X_train,y_train)
```

```python
# Accuracy score on the training data


X_train_eval = lg.predict(X_train)
acc_sc_train = accuracy_score(X_train_eval,y_train)

print('Accuracy score on the training data:', acc_sc_train * 100,'%')
```

```python
# Predict the test data

X_test_predict = lg.predict(X_test)
acc_sc_test = accuracy_score(X_test_predict,y_test)

print('Accuracy score on the test data:', acc_sc_test * 100,'%')
```
